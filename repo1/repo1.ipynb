```python
import numpy as np


def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    y = np.sum(w*x) + b
    # パーセプトロンの出力を返す
    if y <= 0:
        return 0
    if y > 0:
        return 1


def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    y = np.sum(w*x) + b
    # パーセプトロンの出力を返す
    if y <= 0:
        return 0
    if y > 0:
        return 1


def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    y = np.sum(w*x) + b
    # パーセプトロンの出力を返す
    if y <= 0:
        return 0
    if y > 0:
        return 1


def XOR(x1, x2):
    s1 = OR(x1, x2)
    s2 = NAND(x1, x2)
    y = AND(s1, s2)
    # パーセプトロンの出力を返す
    if y <= 0:
        return 0
    if y > 0:
        return 1


if __name__ == '__main__':
    xs_input = [(0, 0), (1, 0), (0, 1), (1, 1)]
    # AND回路の出力
    print('AND')
    for xs in xs_input:
        out = AND(xs[0], xs[1])
        print(str(xs) + " -> " + str(out))
        
    # NAND回路の出力
    print('NAND')
    for xs in xs_input:
        out = NAND(xs[0], xs[1])
        print(str(xs) + " -> " + str(out))
        
    # OR回路の出力
    print('OR')
    for xs in xs_input:
        out = OR(xs[0], xs[1])
        print(str(xs) + " -> " + str(out))
        
    # XOR回路の出力
    print('XOR')
    for xs in xs_input:
        out = XOR(xs[0], xs[1])
        print(str(xs) + " -> " + str(out))
```

# 感想
　NANDのようにANDの出力を反転させたいときは，与える重みw1とw2の符号を反転させればよいことが分かった。x1-x2平面上で条件式の領域を考えると，入力の値と出力される値の対応が分かりやすく感じた。パーセプトロンというものがあるということは知っていたが，重みやバイアスを適切に設定することで論理回路を実装できることは初めて知り，面白く感じた。
